```python
import annotedPlot as ap
import importlib
import spectral.io.envi as envi
import pandas as pd
import matplotlib.pyplot as plt

# Read the spectral library using the spectral package
libVeg = envi.open('UPWINS_4_16_2024.hdr') 
```

```python
feat_params = {'present': {'show_line':True, 'color':'red', 'show_label':True, 'show_wl':True},
                'not_present': {'show_line':True, 'color':'green', 'show_label':False, 'show_wl':False}}
            
aplot = ap.aplot(libVeg.bands.centers, libVeg.spectra[1], libVeg.names[1], 
            plot_gas_trans = True, plot_red_edge = True,
            feat_params=feat_params)
```

```python
# List of specific n values
n_values = [70, 35, 30, 20, 0, 1]

# Loop through specified n values
for n in n_values:
    aplot = ap.aplot(libVeg.bands.centers, libVeg.spectra[n], libVeg.names[n], 
                plot_gas_trans = True, plot_red_edge = True,
                feat_params=feat_params)
```
